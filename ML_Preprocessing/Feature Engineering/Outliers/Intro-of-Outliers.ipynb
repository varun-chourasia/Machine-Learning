# Outliers

Outliers are **data points that are very different** from the rest of the dataset.
Example: Most salaries are ₹20k–₹50k, but one value is ₹5 lakh → that’s an outlier.

## 🔹 When are Outliers Dangerous?
Outliers become risky when they:
- Distort **mean, variance, and model performance**
- Mislead machine learning models (especially linear models / weights based)
- Are caused by **errors** (wrong entry, faulty sensor)

> Note: Not all outliers are bad — sometimes they carry **important insights** (e.g., fraud detection).

## 🔹 How to Treat Outliers?
Depends on context:

- ❌ **Remove** → if it’s clearly an error
- 📉 **Cap (Winsorization)** → limit extreme values
- 🔄 **Transform** → log / square root transformation
- 🤖 **Use robust models** → tree-based models
- ✔ **Keep** → if it's meaningful data

---

## 🔹 One-Line Summary
Outliers are extreme values that can **distort analysis**, so detect them using statistical/visual methods and handle them carefully based on context.
